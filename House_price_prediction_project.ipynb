```python
def pie(column):
    data=df[column].value_counts()
    plt.figure(figsize=(6,6))
    plt.style.use("fivethirtyeight")
    plt.pie(data.values,labels=data.index,autopct="%1.1f%%")
    plt.title(column)
    plt.show()
```

```python
pie("country")
```
